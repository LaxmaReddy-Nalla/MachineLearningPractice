# Manipulating Arrays Shapes

### np.ravel()
### np.flatten()
### np.transpose()
### np.resize()

```python
# np.ravel() it is used convert n dimensional array to 1-dimensional array
a = np.array([[1,3,4],[2,4,5],[23,5,3]])
print(a,'\n')
r = a.ravel()
r
```

```python
# np.flatten()  same as ravel() but continuously assigns novel memory where as ravel might return with a result of an array

a = np.array([[1,3,4],[2,4,5],[23,5,3]],dtype=int)

print(a.flatten())
```

# STACKING IN NUMPY

## Vertical Stacking
## Horizental Stacking
## Depth Stacking 
## Column_Stacking
## Row_Stacking

# Vertical Stacking 
##### In vertical stacking more than 2 matrices are stacked  vertically

```python
a = np.array([[1,2],[3,5]])
b = np.array([[4,4],[43,65]])
print(a,'\n\n',b)
```

```python
np.vstack((a,b))
```

# Horizental Stacking
### In horizental stacking more than 2 matrices are stacked horizentally

```python
a1 = np.array([[1,2],[3,4],[5,3]])
b1 = np.array([[1,34],[2,45],[42,5]])
print(a1,'\n\n',b1)
```

```python
np.hstack((a1,b1))
```

# Depth Stacking 
### In depth stacking 2 or more matrices are combined in depth wise together by added diminsion

```python
a12 = np.array([[1,34],[34,532]])
b12 = np.array([[21,32],[24,53]])
c12 = np.array([[223,42],[54,54]])
print(a12,'\n\n',b12,'\n\n',c12)
```

```python
np.dstack((a12,b12,c12))
```

# Row_stacking 


```python
a = np.array(10*np.random.random((2,2)),dtype=int)
b = np.array(10*np.random.random((2,2)),dtype=int)
print(a,'\n\n',b)
```

```python
np.row_stack((a,b))
np.vstack((a,b))
np.row_stack((a,b))==np.vstack((a,b))
```

# Column Stacking

```python
a = np.array(10*np.random.random((3,3)),dtype=int)
b = np.array(10*np.random.random((3,3)),dtype=int)
print(a,'\n\n',b)
```

```python
np.column_stack((a,b))
```
